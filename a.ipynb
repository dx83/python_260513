```python
import pandas as pd
customer_master = pd.read_csv("http://114.207.245.181:13000/csv/customer_master.csv")
customer_master
```

```python
customer_master.info()
```

```python
customer_master.columns
```

```python
# 딕셔너리 타입으로 앞쪽에 원래:변경
t1 = {'customer_id':'고객아이디', 'customer_name':'고객명',
      'registration_date':'가입일', 'email':'이메일',
      'gender':'성별', 'age':'나이',
      'birth':'생일', 'pref':'지역'}

customer_master = customer_master.rename(columns=t1)

t1 = {'고객아이디':'고객아이디', '고객명':'고객명',
      '등록일자':'등록일', '이메일':'이메일',
      '성별':'성별', '나이':'나이',
      '생일':'생년월일', '지역':'지역'}
customer_master = customer_master.rename(columns=t1)
customer_master
```

```python
customer_master.dtypes
```

```python
# 현재 등록일의 타입이 문자인데 날짜 타입으로 변환하기
customer_master['등록일'] = pd.to_datetime(customer_master['등록일'], format='%Y-%m-%d %H:%M')
customer_master['생년월일'] = pd.to_datetime(customer_master['생년월일'], format='%Y-%m-%d')
```

```python
# 생년월일 컬럼에서 년만 추출해서 새로운 컬럼 "년" 생성
customer_master['년'] = customer_master['생년월일'].dt.year
customer_master['월'] = customer_master['생년월일'].dt.month
customer_master['일'] = customer_master['생년월일'].dt.day
customer_master
```

```python
# 나이 다시 계산하기
customer_master['나이'] = 2026 - customer_master['년'] + 1
customer_master
```

```python
bins = [0, 9, 19, 29, 39, 49, 59, 69, 79, 150]
labels = ["0대", "10대", "20대", "30대", "40대", "50대", "60대", "70대", "초고령"]
customer_master['연령대'] = pd.cut(customer_master['나이'], bins=bins, labels=labels, right=True)
customer_master
```

```python
# 데이터프레임 customer_master 객체를 파일로 저장하기
customer_master.to_pickle('./pkl/customer_master.pkl')
```

```python
# 거래내역1 읽기
transaction_1 = pd.read_csv('http://114.207.245.181:13000/csv/transaction_1.csv')
transaction_1.head(3)
```

```python
# 거래내역2 읽기
transaction_2 = pd.read_csv('http://114.207.245.181:13000/csv/transaction_2.csv')
transaction_2.head(3)
```

```python
# transaction_1에 transaction_2를 붙이기
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction
```

```python
transaction.columns
```

```python
#컬럼명 변경하기
t1 = { 'transaction_id':'거래코드', 'price':'결제금액',
      'payment_date':'결제일', 'customer_id':'고객아이디' }
transaction = transaction.rename(columns=t1)
transaction
```

```python
# 타입확인하기
transaction.dtypes
```

```python
# 결제일을 날짜 타입으로 변환
transaction['결제일'] = pd.to_datetime(transaction['결제일'], format='%Y-%m-%d %H:%M:%S')
```

```python
# transaction을 pkl로 저장
transaction.to_pickle('./pkl/transaction_master.pkl')
```
