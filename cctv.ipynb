```python
import pandas as pd
df= pd.read_csv("http://114.207.245.181:13000/csv/cctv01.csv")
df
```

```python
# 인덱스 0인것 지우기
df = df.drop(index=[0])
```

```python
df
```

```python
df.dtypes
```

```python
df['총계'] = df['총계'].str.replace(",","")
df['총계'] = df['총계'].astype("int64")
df
```

```python
df['2011년 이전'] = df['2011년 이전'].str.replace(",","")
df['2011년 이전'] = df['2011년 이전'].astype("int64")
df['2011년'] = df['2011년'].str.replace(",","")
df['2011년'] = df['2011년'].astype("int64")
df['2012년'] = df['2012년'].str.replace(",","")
df['2012년'] = df['2012년'].astype("int64")
df['2014년'] = df['2014년'].str.replace(",","")
df['2014년'] = df['2014년'].astype("int64")
df['2015년'] = df['2015년'].str.replace(",","")
df['2015년'] = df['2015년'].astype("int64")
df['2016년'] = df['2016년'].str.replace(",","")
df['2016년'] = df['2016년'].astype("int64")
df['2017년'] = df['2017년'].str.replace(",","")
df['2017년'] = df['2017년'].astype("int64")
df['2018년'] = df['2018년'].str.replace(",","")
df['2018년'] = df['2018년'].astype("int64")
df['2019년'] = df['2019년'].str.replace(",","")
df['2019년'] = df['2019년'].astype("int64")
df['2020년'] = df['2020년'].str.replace(",","")
df['2020년'] = df['2020년'].astype("int64")
```

```python
df
```

```python
# 컬럼의 위치 1번부터 끝까지를 한번에 변경하기
df[df.columns[1:]] = df[df.columns[1:]].replace(",", "", regex=True).astype("int64")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('font', family='Gulim')

plt.figure(figsize=( 15, 5))
sns.barplot(data=df, x='구분', y='총계')
plt.xticks(rotation=45) # x축 항목명 회전
plt.show()
```
