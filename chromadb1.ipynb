```python
# crewai 역할 기반 작업 분담
from langchain_ollama import OllamaLLM
from crewai import LLM

ollama_llm = LLM(
    model="ollama/exaone-3.5:latest",
    base_url = "http://192.168.0.12:11435",
    temperature=0.3     # 0.3은 코드 요약 분석용
)
```

```python
from crewai import Agent

researcher = Agent(
    role="AI Researcher",
    goal="주어진 주제에 대해서 개념을 조사한다",
    backstory="AI 논문과 기술 문서를 분석하는 전문가",
    llm=ollama_llm,
    verbose=True
)

writer = Agent(
    role="Technical Writer",
    goal="조사 결과를 개발자 중심으로 정리한다.",
    backstory="개발자를 위한 기술 문서를 잘 쓰는 작가",
    llm=ollama_llm,
    verbose=True
)
```

```python
# Task 정의
from crewai import Task

research_task = Task(
    description="CrewAI와 Ollama의 역할과 장점을 조사해라",
    expected_output="CrewAI와 Ollama의 핵식 역할과 장점을 요약한 정리 문서",
    agent=researcher
)

writer_task = Task(
    description="조사결과를 개발자용으로 쉽게 정리해라",
    expected_output="비 전공자도 이해 할 수 있는 쉬운 설명 문서",
    agent=writer
)
```

```python
# Agent + Task 묶기
from crewai import Crew

crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, writer_task],
    verbose=True
)
```

```python
result = await crew.kickoff_async()
result
```
