```python
import time
import requests
import os

from urllib.parse import urlparse

from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By

options = Options()

# options.add_argument("--headless")
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36 Edg/123.0.0.0"
options.add_argument(f"user-agent={user_agent}")

options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_argument("--lang=ko-KR,ko")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

driver = webdriver.Edge(service=Service("./driver/msedgedriver.exe"), options=options)

try :
    url = 'https://store.kyobobook.co.kr/bestseller/online/daily'
    driver.get(url)
    time.sleep(1.5)

    for i in range(1, 3):
        for j in range(1, 11):
            xpath = f'/html/body/div[2]/main/section/div/div/section/ol[{i}]/li[{j}]/div/div[2]/div[2]/a'
            title = driver.find_element(By.XPATH, xpath)
            print(title.text)

            xpath = f'/html/body/div[2]/main/section/div/div/section/ol[{i}]/li[{j}]/div/div[2]/div[1]/a/div/img'
            img = driver.find_element(By.XPATH, xpath)
            src = img.get_attribute('src')
            print(src)

            # src에서 파일명만 추출
            parsed = urlparse(src)
            filename = os.path.basename(parsed.path)
            print(filename)

            response = requests.get(src)
            if response.status_code == 200:
                with open(f'./image/{filename}', 'wb') as f:
                    f.write(response.content)

except Exception as e:
    print(e)
finally:
    driver.quit()
```

```python
import time
import requests
import os
import re

from urllib.parse import urlparse

from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By

options = Options()

# options.add_argument("--headless")
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36 Edg/123.0.0.0"
options.add_argument(f"user-agent={user_agent}")

options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_argument("--lang=ko-KR,ko")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

driver = webdriver.Edge(service=Service("./driver/msedgedriver.exe"), options=options)

try :
    url = 'https://shop.teacherville.co.kr/goods/best'
    driver.get(url)
    time.sleep(1)

    for i in range(1, 101):
        xpath = f'//*[@id="goods_best_common"]/li[{i}]/div[1]'
        div   = driver.find_element(By.XPATH, xpath)
        style = div.get_attribute('style')
        match = re.search(r'url\(["\']?([^"\')]+)["\']?\)', style)

        if match:
            print(f'https://shop.teacherville.co.kr/{match.group(1)}')

            src = f'https://shop.teacherville.co.kr/{match.group(1)}'
            parsed = urlparse(src)
            filename = os.path.basename(parsed.path)
            print(filename)

            response = requests.get(src)
            if (response.status_code == 200):
                with open(f'./image/{filename}', 'wb') as f:
                    f.write(response.content)

except Exception as e:
    print(e)

finally:
    driver.quit()
```
