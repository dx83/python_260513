```python
import pandas as pd
df = pd.read_csv("http://114.207.245.181:13000/csv/fish01.csv")
```

```python
df.columns
```

```python
# 컬럼명 변경하기, inplace=True를 사용해야 실제 변경됨
t1 = {'Species':'종류', 'Weight':'무게', 'Length':'길이', 'Diagonal':'대각선', 'Height':'높이', 'Width':'넓이'}
df.rename(columns=t1, inplace=True)
```

```python
df['종류'].unique()
```

```python
df['종류'].value_counts()
```

```python
t1 = {'Bream':'도미', 'Roach':'로치', 'Whitefish':'화이트피시', 'Parkki':'파르키', 'Perch':'농어', 'Pike':'파이크', 'Smelt':'빙어'}
df['종류'] = df['종류'].map(t1)
df
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
rc('font', family='Gulim')
plt.figure(figsize=(10, 10))
sns.pairplot(df[['무게','길이','대각선','높이','넓이']])
plt.show()
```

```python
# 농어의 길이와 무게만 가져와서 df1에 보관
c1 = (df['종류'] == '농어')
df1 = df.loc[c1, ['길이','무게']]
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('font', family='Gulim')
plt.figure(figsize=(7, 7))
sns.scatterplot(data=df1, x='길이', y='무게')
plt.show()
```

```python
x = df1[['길이']].values
x
```

```python
y = df1['무게'].values
y
```

```python
x.shape, y.shape
```

```python
# 데이터 분할, 학습용 80%, 검증용 20%
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
# 선형회귀모델
lr = LinearRegression()
# 데이터 학습
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train), lr.score(x_test, y_test)
```

```python
# 학습한 모델 저장하기
import joblib
joblib.dump(lr, './pkl/20260519_model.joblib')
```

```python
# 모델 읽기, 백엔드, 추론시스템
import joblib
lr = joblib.load('./pkl/20260519_model.joblib')
lr.score(x_train, y_train), lr.score(x_test, y_test)
```

```python
c1 = (df['종류'] == '농어')

df2 = df.loc[c1, ['길이','무게','넓이','높이']]
df2
```

```python
x = df2[['길이','넓이','높이']].values
y = df2['무게'].values

x.shape, y.shape
```

```python
# 데이터 분할, 학습용 80%, 검증용 20%
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_train, y_train), lr.score(x_test, y_test)
```

```python
lr.coef_, lr.intercept_
```

```python
# y = a1 * 가중치1 + a2 * 가중치2 + a2 * 가중치3 + 기울기
x_test[0], y_test[0]
```

```python
sample = np.array([[27.5,  4.335, 7.0516]])
lr.predict(sample)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3, include_bias=False)
poly.fit(x_train)

x_train_poly = poly.transform(x_train)
x_test_poly = poly.transform(x_test)

x_train_poly.shape, x_test_poly.shape
```

```python
lr.fit(x_train_poly, y_train)
lr.score(x_train_poly, y_train), lr.score(x_test_poly, y_test)
```

```python
# 릿지 읿부 가중치 크기를 최소화
# 라쏘 일부 가중치를 0으로 만듬
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(x_train_poly, y_train)
lasso.score(x_train_poly, y_train), lasso.score(x_test_poly, y_test)
```

```python
lasso.coef_
```

```python
# 학습한 모델 저장하기
import joblib
joblib.dump(lasso, './pkl/20260519_model1.joblib')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Lasso

# pipeline 구성
pipe = Pipeline([
    ('poly', PolynomialFeatures(degree=5, include_bias=False)),
    ('lasso', Lasso(alpha=1.0))
])

# 특성을 놀리기전 3개의 특성
pipe.fit(x_train, y_train)
```

```python
import joblib
joblib.dump(pipe, './pkl/20260519_lasso.joblib')
```
