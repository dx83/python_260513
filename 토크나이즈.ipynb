```python
# 단어 사전 만들기
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer

text = ["오늘 라일락 메뉴가 어떻게 되나요", "자세한 설명 부탁드립니다", "해당 장비를 정지합니다"]

# 사전에 없는 단어
tokenizer = Tokenizer(oov_token="<UNK>")
tokenizer.fit_on_texts(text)

voca_size = len(tokenizer.word_counts) + 1

tokenizer.word_index, voca_size
```

```python
# 단어사전을 기반으로 문자를 숫자로 변경하기
input = ["오늘 라일락 메뉴는 오마카세가 아닙니다", "설명 짧고 굵게 해주세요", "해당 장비를 정지합니다"]

input_token = tokenizer.texts_to_sequences(input)
input_token
```

```python
# 입력한 문장에 따라서 길이가 다름
from tensorflow.keras.preprocessing.sequence import pad_sequences

input_pad = pad_sequences(input_token, maxlen=5, padding="post") # pre(앞에 0) 또는 post(뒤에 0)
input_pad
```

```python
# 단어횟수
text1 =["오늘 설명 해주세요", "해당 장비를 설명 해당 "]
tokenizer.fit_on_texts(text1)
print(tokenizer.word_index)
tokenizer.texts_to_matrix(text1, mode="count")  # mode="count", binary(0,1)
```
