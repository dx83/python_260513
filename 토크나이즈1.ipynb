```python
import pandas as pd

df = pd.read_csv("http://114.207.245.181:13000/csv/language.csv")
df.head()
```

```python
df['label_name'].value_counts()
```

```python
texts = df['text'].values
texts
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

voca_size = 1000

tokenizer = Tokenizer(num_words=voca_size, oov_token="<UMK>")
tokenizer.fit_on_texts(texts)

tokenizer.word_index
```

```python
# 단어 빈도수 구하기
x = tokenizer.texts_to_matrix(texts, mode="count")
y = df[['label']].values
x.shape, y.shape
```

```python
# 8:2로 나누기
from sklearn.model_selection import train_test_split

# stratify=y는 y클래스를 일정한 비율로 나누기
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123, stratify=y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
    Input(shape=(1000,)),
    Dense(units=128, activation="relu"),
    Dense(units=64, activation="relu"),
    Dense(units=4, activation="softmax")
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=400, validation_data=(x_test, y_test), verbose=2)
```

```python
label = {0:"Java", 1:"Python", 2:"Javascript", 3:"PHP"}

sample = [
    "I enjoy building enterprise applications with Spring Boot and Java",
    "My favorite language for data analysis and machine learning is Python.",
    "I use JavaScript to create interactive web pages and frontend applications",
    "JavaScript is essential for modern web development and browser-based applications.",
    "PHP powers many content management systems and web backends.",
    "I prefer Java beacause of its strong type system and JVM ecosystem.",
    "Python is my first choice when working with AI and data science projects."
]

sample_text = tokenizer.texts_to_matrix(sample, mode="count")
pred = model.predict(sample_text)
pred_labels = pred.argmax(axis=1)
for s, p in zip(sample, pred_labels):
    print(f"{s} => {label[p]}")
```
