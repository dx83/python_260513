```python
import pandas as pd
df = pd.read_pickle('./pkl/student.pkl')
df
```

```python
t1 = { "Freshman":"1학년", "Sophomore":"2학년", "Junior":"3학년", "Senior":"4학년" }
df['학년'] = df['학년'].map(t1)
df
```

```python
df1 = pd.read_csv("http://114.207.245.181:13000/csv/score_info.csv")
df1
```

```python
t1 = { "Student_ID":"학생아이디", "Math":"수학", "Science":"과학", 
      "English":"영어", "History":"역사", "Physical_Education":"체육" }
df1 = df1.rename(columns=t1)
df1
```

```python
df2 = pd.merge(df, df1, on="학생아이디")
df2.head(3)
```

```python
df2['합계'] = df2['수학'] + df2['과학'] + df2['영어'] + df2['역사'] + df2['체육']
df2
```

```python
df2['평균'] = df2['합계'] / 5
df2
```

```python
# min은 예시) 1 2 2 4
df2['석차'] = df2['합계'].rank(method="min", ascending=False).astype(int)
df2
```

```python
df2.head()
```

```python
# 학년별 합계 평균 구하기
# groupby([학년])[[]].연산(sum, mean, min, max, ...)
df2.groupby(['학년'])[['합계']].mean().round(2)
```

```python
df2.groupby(['학년'])[['수학', '영어']].agg(["mean", "sum", "min", "max"])
```
